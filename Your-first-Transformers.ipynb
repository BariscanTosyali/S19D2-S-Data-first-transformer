### Basic Sentiment : 😀 /  😕 / 😠 / 😟

Bunu göz önünde bulundurarak, bir model belirtmeden duygu analizi için bir pipeline oluşturun ve "Transformers are awesome!" cümlesiyle bu modeli test etmeye çalışın. Diğer cümleleri de denemekten çekinmeyin.

```python
from transformers import pipeline

# Herhangi bir model belirtmeden duygu analizi pipeline'ı oluşturuyoruz
basic_classifier = pipeline("sentiment-analysis")

# "Transformers are awesome!" cümlesini test edelim
result = basic_classifier("Transformers are awesome!")
print(f"Test 1: {result}")

# Birkaç farklı duygu içeren cümle daha deneyelim
sentences = [
    "I am having a wonderful day!",
    "The weather is a bit gloomy today.",
    "I'm not sure how I feel about this new update."
]

for sentence in sentences:
    res = basic_classifier(sentence)
    print(f"Cümle: {sentence} -> Sonuç: {res}")
```

### Translation ✍️

Dil temamıza sadık kalalım ve bizim için bazı cümleleri çevirme görevlerini yerine getirebilecek bir model bulmaya çalışalım. Helsinki Üniversitesi'nin `opus-mt` projesi, HuggingFace'de inanılmaz derecede aktif olup, birçok farklı küresel dil için çeviri sürecini demokratikleştirmek üzere tasarlanmış modeller oluşturmakta ve sürdürmektedir. Şunu uygulamayı deneyin `"Helsinki-NLP/opus-mt-<source-language>-<destination-language>"` tiki dil arasında çeviri yapıp yapamayacağınızı görmek için (örneğin İngilizce'den İspanyolca'ya).

```python
from transformers import pipeline

# İngilizce'den (en) İspanyolca'ya (es) bir çeviri pipeline'ı oluşturuyoruz
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es")

# Çevirmek istediğimiz cümle
text_to_translate = "HuggingFace makes Natural Language Processing accessible for everyone."

# Çeviriyi gerçekleştir
translation_result = translator(text_to_translate)

print(f"Orijinal: {text_to_translate}")
print(f"İspanyolca Çeviri: {translation_result[0]['translation_text']}")
```

### Özetleme

Bir başka gerçekten yararlı NLP görevi, büyük miktarda bilgiyi çok az sayıda kelimeyle özetlemektir. BART, özetleme gibi görevlerde iyi performans gösteren bir modeldir; derslerde kısaca gördüğünüz iki modelin birleşiminden oluşur: BERT modeli ve otogresif stil GPT modeli. Bu model hakkında daha fazla bilgi için [bu bağlantıya](https://www.projectpro.io/article/transformers-bart-model-explained/553) göz atın. 

BART modelleri oldukça büyük olabileceğinden, HuggingFace'de mevcut en küçük damıtma modellerinden biri olan `distilbart-xsum-12-6` modelini bulmaya çalışın (damıtma hakkında daha sonra daha fazla konuşacağız!). Bu modeli bir `“özetleme”` boru hattına entegre edin, ardından bir metin alın (örneğin, [bir BBC makalesini](https://www.bbc.com/news/topics/cx2pk70323et) kopyalayıp yapıştırarak) ve boru hattınızla özetleyin!

Not: Bağlam pencerelerine dikkat etmelisiniz - burada, girişinizin model için çok uzun olması nedeniyle bir sorunla karşılaşabilirsiniz!

```python
from transformers import pipeline

# Özetleme için 'summarization' pipeline'ı oluşturuyoruz
# distilbart-xsum-12-6 modeli oldukça hızlı ve verimlidir.
summarizer = pipeline("summarization", model="sshleifer/distilbart-xsum-12-6")

# Örnek bir makale metni (Burayı dilediğin metinle güncelleyebilirsin)
text = """
The Amazon rainforest, also known in English as Amazonia or the Amazon Jungle, is a moist broadleaf forest 
that covers most of the Amazon basin of South America. This basin encompasses 7,000,000 square kilometres, 
of which 5,500,000 square kilometres are covered by the rainforest. This region includes territory 
belonging to nine nations. The majority of the forest is contained within Brazil, with 60% of the 
rainforest, followed by Peru with 13%, Colombia with 10%, and with minor amounts in Venezuela, 
Ecuador, Bolivia, Guyana, Suriname and French Guiana.
"""

# Özetleme işlemini başlatıyoruz
# max_length: Özetin maksimum uzunluğu
# min_length: Özetin minimum uzunluğu
summary = summarizer(text, max_length=50, min_length=10, do_sample=False)

print("--- Orijinal Metin ---")
print(text)
print("\n--- Özet ---")
print(summary[0]['summary_text'])
```

### Daha ileri gitmek: Soru Cevaplama 🔍

Özetin ötesine geçmek istersek ne yapmalıyız? Belirli bir veri kümesi hakkında sezgisel bir şekilde sorular sormak istersek? Bunun için de bir model var! (Oldukça küçük) `roberta-base-squad2` modelini kullanın. Bu model, sağlanan `context` (sağlayacağınız bir metin) hakkında `soru`ları yanıtlayabilen soru-cevap çiftleri üzerinde eğitilmiştir. Belgeleri inceleyin. [burada](https://huggingface.co/deepset/roberta-base-squad2?context=The+Amazon+rainforest+%28Portuguese%3A+Floresta+Amaz%C3%B4nica+or+Amaz%C3%B4nia%3B+Spanish%3A+Selva+Amaz%C3%B3nica%2C+Amazon%C3%ADa+or+usually+Amazonia%3B+French%3A+For%C3%AAt+amazonienne%3B+Dutch%3A+Amazoneregenwoud%29%2C+also+known+in+English+as+Amazonia+or+the+Amazon+Jungle%2C+is+a+moist+broadleaf+forest+that+covers+most+of+the+Amazon+basin+of+South+America.+This+basin+encompasses+7%2C000%2C000+square+kilometres+%282%2C700%2C000+sq+mi%29%2C+of+which+5%2C500%2C000+square+kilometres+%282%2C100%2C000+sq+mi%29+are+covered+by+the+rainforest.+This+region+includes+territory+belonging+to+nine+nations.+The+majority+of+the+forest+is+contained+within+Brazil%2C+with+60%25+of+the+rainforest%2C+followed+by+Peru+with+13%25%2C+Colombia+with+10%25%2C+and+with+minor+amounts+in+Venezuela%2C+Ecuador%2C+Bolivia%2C+Guyana%2C+Suriname+and+French+Guiana.+States+or+departments+in+four+nations+contain+%22Amazonas%22+in+their+names.+The+Amazon+represents+over+half+of+the+planet%27s+remaining+rainforests%2C+and+comprises+the+largest+and+most+biodiverse+tract+of+tropical+rainforest+in+the+world%2C+with+an+estimated+390+billion+individual+trees+divided+into+16%2C000+species.&question=How+many+species+are+in+the+Amazon%3F).

Nasıl yapılacağını biliyorsunuz: `roberta-base-squad2` modeli ile bir `“soru-cevap”` pipeline oluşturun, ardından daha önce seçtiğiniz `makaleyi` bağlam olarak ekleyin ve bununla ilgili bir `soru` sorun.

```python
from transformers import pipeline

# Soru-cevap için 'question-answering' pipeline'ı oluşturuyoruz
# 'deepset/roberta-base-squad2' modeli bu iş için oldukça optimize edilmiştir.
qa_model = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Bir önceki adımda kullandığımız Amazon makalesini bağlam (context) olarak kullanalım
context_text = """
The Amazon rainforest, also known in English as Amazonia or the Amazon Jungle, is a moist broadleaf forest 
that covers most of the Amazon basin of South America. This basin encompasses 7,000,000 square kilometres, 
of which 5,500,000 square kilometres are covered by the rainforest. This region includes territory 
belonging to nine nations. The majority of the forest is contained within Brazil, with 60% of the 
rainforest, followed by Peru with 13%, Colombia with 10%, and with minor amounts in Venezuela, 
Ecuador, Bolivia, Guyana, Suriname and French Guiana.
"""

# Sorunu sor
question = "How many nations does the Amazon region include territory from?"

# Modeli çalıştır
result = qa_model(question=question, context=context_text)

print(f"Soru: {question}")
print(f"Cevap: {result['answer']}")
print(f"Güven Skoru: {result['score']:.4f}")
```

Bu işlem tamamlandıktan sonra, bir `“document-question-answering”` pipeline oluşturacağız. Bunun için bir modele ihtiyacımız olacak, bu nedenle HuggingFace'de `layoutlm-invoices` modelini arayın. Ardından bununla ilgili sorular sormaya çalışın.   [`receipt.webp`](https://drive.google.com/file/d/1slICDEjI1AXuWXybfSBIt5cCzteFgb3C/view?usp=sharing) (görüntüyü veri klasörünüze indirin veya modeli çağırdığınızda URL'yi doğrudan modele aktarın).
```python
# Mac için, yorum satırını kaldırın:
# !brew install tesseract

# Linux veya Colab vb. için, şu satırların yorum satırını kaldırın:
# !sudo apt install tesseract-ocr  
# !sudo apt install libtesseract-dev

# Ardından çekirdeğinizi yeniden başlatın ve deneyin!
```

```python
from transformers import pipeline

# 1. 'document-question-answering' pipeline'ı oluşturalım
# 'impira/layoutlm-document-qa' modeli faturalar için oldukça yeteneklidir.
document_qa = pipeline("document-question-answering", model="impira/layoutlm-document-qa")

# 2. Makbuz görselinin URL'si (veya yerel yolu: "data/receipt.webp")
image_url = "https://huggingface.co/spaces/impira/docquery/resolve/2359223c1c8f1e680a6b5790757c9636ef7c0c17/receipt.png"

# 3. Sorularımızı soralım
questions = [
    "What is the total amount?",
    "What is the price of the eggs?",
    "How much is the sales tax?"
]

print("Görsel analiz ediliyor...")

for question in questions:
    result = document_qa(image=image_url, question=question)
    print(f"\nSoru: {question}")
    print(f"Cevap: {result[0]['answer']} (Güven: {result[0]['score']:.4f})")
```
